IMPORTING LIBRARIES

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import r2_score
```

LOADING DATASET

```python
df=pd.read_csv('Salary_dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

CHECKING FOR NULL VALUES

```python
df.isnull().sum()
```

```python
plt.ylabel('Years of experience')
plt.xlabel('Salary')
plt.scatter( df.Salary,df.YearsExperience ,marker='+', color='r' )
plt.show()
```

OBJECT CREATION

```python
reg=linear_model.LinearRegression()
reg.fit(df[['Salary']],df.YearsExperience)
```

PREDICTING VALUES

```python
reg.predict([[40000]])
```

DATA VISUALIZATION

```python
plt.ylabel('Years of experience')
plt.xlabel('Salary')
plt.scatter( df.Salary,df.YearsExperience ,marker='+', color='r' )
plt.plot( df.Salary,reg.predict(df[['Salary']]))
plt.show()
```
